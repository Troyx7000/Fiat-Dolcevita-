# Visualizations of Car Brand Market Share in the UK

These visualizations were sourced from a literature review focused on analyzing competitors in the market. The data was extracted from the primary source of UK vehicle information: the Society of Motor Manufacturers and Traders (SMMT) car registration data, available at https://www.smmt.co.uk/vehicle-data/car-registrations/.

```python
import pandas as pd
```

```python
# Load the dataset
file_path = r'C:\Users\melis\OneDrive\Desktop\ARP 3\car_sales_data_SMMT.csv'
carsales_df = pd.read_csv(file_path)
```

```python
df = carsales_df
```

```python
import matplotlib.pyplot as plt

# Data for plotting
brands = df['MARQUE']
market_share_2024 = df['2024 % Market share'].replace("", 0).astype(float)  # Replace empty strings with 0 and convert to float

# Plotting
plt.figure(figsize=(10, 8))
plt.barh(brands, market_share_2024, color='skyblue')
plt.xlabel('Market Share (%)')
plt.ylabel('Car Brands')
plt.title('Market Share of Car Brands in 2024')
plt.gca().invert_yaxis()  # Invert y-axis to have the largest market share at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for plotting
brands = df['MARQUE']
market_share_2024 = df['2024 % Market share'].replace("", 0).astype(float)  # Replace empty strings with 0 and convert to float

# Combine the data into a new DataFrame and sort by market share
top_10_df = df[['MARQUE', '2024 % Market share']].replace("", 0).astype({'2024 % Market share': float})
top_10_df = top_10_df.sort_values(by='2024 % Market share', ascending=False).head(10)

# Data for top 10
top_10_brands = top_10_df['MARQUE']
top_10_market_share = top_10_df['2024 % Market share']

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_10_brands, top_10_market_share, color='coral')
plt.xlabel('Car Brands')
plt.ylabel('Market Share (%)')
plt.title('Top 10 Car Brands by Market Share in 2024')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sort data by market share and separate top 10 brands
sorted_df = df[['MARQUE', '2024 % Market share']].replace("", 0).astype({'2024 % Market share': float}).sort_values(by='2024 % Market share', ascending=False)

# Get top 10 brands
top_10 = sorted_df.head(20)

# Calculate the rest as "Others"
others_market_share = sorted_df.iloc[5:]['2024 % Market share'].sum()

# Combine top 10 with "Others"
labels = top_10['MARQUE'].tolist() + ['Others']
sizes = top_10['2024 % Market share'].tolist() + [others_market_share]

# Plotting Pie Chart
plt.figure(figsize=(10, 10))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Market Share of Top 10 Car Brands vs. Others in 2024')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter data for Fiat only
fiat_market_share = df[df['MARQUE'] == 'Fiat']['2024 % Market share'].values[0]

# Data for Pie Chart
labels = ['Fiat', 'Other Brands']
sizes = [fiat_market_share, 100 - fiat_market_share]
colors = ['#FF9999', '#66B2FF']

# Plotting Pie Chart for Fiat
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=(0.1, 0))  # Explode Fiat slice
plt.title('Market Share of Fiat vs. Other Brands in 2024')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the change in market share between 2023 and 2024
df['Change in Market Share (%)'] = df['2024 % Market share'].replace("", 0).astype(float) - df['2023 % Market share'].replace("", 0).astype(float)

# Sort data by change in market share
sorted_change_df = df.sort_values(by='Change in Market Share (%)', ascending=False)

# Plotting the change in market share
plt.figure(figsize=(12, 10))
plt.barh(sorted_change_df['MARQUE'], sorted_change_df['Change in Market Share (%)'], color='lightgreen')
plt.xlabel('Change in Market Share (%)')
plt.ylabel('Car Brands')
plt.title('Change in Market Share from 2023 to 2024')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
